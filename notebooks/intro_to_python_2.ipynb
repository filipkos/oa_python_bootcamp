```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

### Plotting simple function

```python
x_array = np.linspace(0, 2*np.pi, 100)
y_array = np.sin(x_array)
```

```python
plt.plot(x_array, y_array, color="green", linewidth='21')
plt.xlabel("X")
plt.ylabel("Sin(X)")
plt.tight_layout()
```
